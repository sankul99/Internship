```python
driver = webdriver.Chrome("chromedriver.exe")

driver.get("https://www.digit.in/top-products/best-gaming-laptops-40.html")

```

```python
Brands=[]
Products_Description=[]
Specification=[]
Price=[]
```

```python
brand =driver.find_elements(By.XPATH,"//div[@class='TopNumbeHeading active sticky-footer']")
len(brand)
```

```python
for i in brand:
   
    Brands.append(str(i.text).replace("\n",""))
    Brands

sp=driver.find_elements(By.XPATH,"//div[@class='Specs-Wrap']")
len(sp)

for i in sp:
   
    Specification.append(str(i.text).replace("\n",""))
Specification
```

```python
des=driver.find_elements(By.XPATH,"//div[@class='Section-center']")
len(des)

for i in des:
   
    Products_Description.append(str(i.text).replace("\n",""))
Products_Description
```

```python

pri=driver.find_elements(By.XPATH,"//td[@class='smprice']")
len(pri)

for i in pri:
   
    Price.append(str(i.text).replace("\n",""))
Price
```

```python
digit_lap=pd.DataFrame([])
digit_lap['Brands']=Brands[0:10]
digit_lap['Price']=Price[0:10]
digit_lap['Specification']=Specification[0:10]
digit_lap['Description']=Products_Description[0:10]
digit_lap
```

QUESTION 8

```python
driver = webdriver.Chrome("chromedriver.exe")

driver.get("https://www.forbes.com/billionaires/")

```

```python
Rank = []
Name = []
Citizenship = []
Age = []
Networth = []
Source = []
Industry = []
```

```python
#Getting names first 

name_tag = driver.find_elements(By.XPATH,'//div[@class="personName"]')
for i in name_tag:
    Name.append(i.text)
    
#Getting Ranks now 

rank_tag = driver.find_elements(By.XPATH,'//div[@class="rank"]')
for i in rank_tag:
    Rank.append(i.text)
    
#Citizenship 

cit = driver.find_elements(By.XPATH,'//div[@class="countryOfCitizenship"]')
for i in cit:
    Citizenship.append(i.text)
    
#Age 
 
age = driver.find_elements(By.XPATH,'//div[@class="age"]')
for i in age:
    Age.append(i.text)
    
#Networth 

net = driver.find_elements(By.XPATH,'//div[@class="netWorth"]')
for i in net:
    Networth.append(i.text)
    
#industry

ind = driver.find_elements(By.XPATH,'//div[@class="category"]')
for i in ind:
    Industry.append(i.text)
    
#Source

sou = driver.find_elements(By.XPATH,'//div[@class="expand-row__icon-container"]')
for i in sou:
    Source.append(i.text)
```

```python
Industry
```

```python
forbes_list=pd.DataFrame()
forbes_list['NAME']= Name
forbes_list['RANK']= Rank
forbes_list['CITIZENSHIP']= Citizenship
forbes_list['AGE']= Age
forbes_list['NETWORTH'] = Networth
forbes_list['INDUSTRY'] = Industry
forbes_list['SOURCE']= Source
```

```python
driver = webdriver.Chrome("chromedriver.exe")

driver.get("https://www.hostelworld.com/")
```

```python
search = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div/div/div[4]/div/div[2]/div/div[1]/div/input')
search.send_keys("London")

button = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div/div/div[4]/div/div[2]/div/div[5]')
button.click()
```

```python
hostel_name = []
distance = []
ratings = []
tot_review = []
ove_review = []
privates = []
dorms = []
facilities = []
description = []
```

```python
name_tag = driver.find_elements(By.XPATH,'//h2[@class="title title-6"]')
for i in name_tag[0:29]:
    hostel_name.append(i.text)
    
distance_tag = driver.find_elements(By.XPATH,'//div[@class="subtitle body-3"]')
for i in distance_tag[0:29]:
    distance.append(i.text)
    
ratings_tag = driver.find_elements(By.XPATH,'//div[@class="score orange big"]')
for i in ratings_tag[0:29]:
    ratings.append(i.text)
    
tot_tag = driver.find_elements(By.XPATH,'//div[@class="reviews"]')
for i in tot_tag[0:29]:
    tot_review.append(i.text)
    
over_tag = driver.find_elements(By.XPATH,'//div[@class="keyword"]')
for i in over_tag[0:29]:
    ove_review.append(i.text)
    
privates_tag = driver.find_elements(By.XPATH,'//div[@class="price title-5"]')
for i in privates_tag[0:29]:
    privates.append(i.text)
    
dorms_tag = driver.find_elements(By.XPATH,'//div[@class="price title-5"]')
for i in dorms_tag[0:29]:
    dorms.append(i.text)
    
faci_tag = driver.find_elements(By.XPATH,'//div[@class="facilities-label facilities"]')
for i in faci_tag[0:29]:
    facilities.append(i.text)
    
```

```python
hostel_world=pd.DataFrame()
hostel_world['Hostel Name']= hostel_name
hostel_world['Distance form City Centre']= distance
hostel_world['Total Number Of Reviews']= tot_review
hostel_world['Overall Reviews']= ove_review
hostel_world['Private Rooms'] = privates
hostel_world['Dorm Rooms'] = dorms
hostel_world['Facilities']= facilities
```
