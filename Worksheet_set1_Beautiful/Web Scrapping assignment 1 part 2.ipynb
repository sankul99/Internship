5. Write a python program to scrape cricket rankings from icc-cricket.com

a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating

```python
page = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page
```

```python
odi = BeautifulSoup(page.content)
odi
```

```python
odi_data = odi.find_all("div", class_="rankings-block__container full rankings-table")
odi_data
```

```python
for data in odi_data:
    team = data.find_all("span", class_ ="u-hide-phablet")
    m_p = data.find_all("td", class_ = "table-body__cell u-center-text")
    rating = data.find_all("td", class_ = "table-body__cell u-text-right rating")
```

```python
matches = []
points = []
odi_df = pd.DataFrame()
teams = []
ratings = []
    
```

```python
for i in range(20):
    if i%2 == 0:
        matches.append(m_p[i].text.strip())
    else:
        points.append(m_p[i].text.strip())

for i in range(10):
    teams.append(team[i].text.strip())
    ratings.append(rating[i].text.strip())
```

```python
odi_df['Team'] = teams
odi_df['Matches'] = matches
odi_df['Points'] = points
odi_df['Rating'] = ratings
odi_df.head(10)
```

7. Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world :
i) Headline
ii) Time
iii) News Link

```python
page = requests.get('https://www.cnbc.com/world/?region=world')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
news_data = soup.find_all('li', class_='LatestNews-item')
news_data
```

```python
time = []
links = []
headline = []
```

```python
for data in news_data:
    t = data.find('time',class_ = 'LatestNews-timestamp').text.strip()
    time.append(t)
    for link in data.find_all('a'):
        l = link.get('href') 
        h = link.get('title')
        if l != None and h != None:
            links.append(l)
            headline.append(h)
            
        
```

```python
news_df = pd.DataFrame()
news_df['Headline'] = headline
news_df['Links'] = links
news_df['Time'] = time

news_df.head()
```

8. Write a python program to scrape the details of most downloaded articles from AI in last 90 days. 
https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles
Scrape below mentioned details :
i) Paper Title 
ii) Authors
iii) Published Date 
iv) Paper URL

```python
page = requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
page
```

```python
soap = BeautifulSoup(page.content)
soap
```

```python
paper_data = soap.find_all('li',class_='sc-9zxyh7-1 sc-9zxyh7-2 kOEIEO hvoVxs')
paper_data
```

```python
title = []
url = []
authors = []
date = []
```

```python
for data in paper_data:
    for d1 in data.find_all('a'):
        url.append(d1.get('href'))
        title.append(d1.find('h2',class_ = 'sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg').text.strip())
    for d2 in data.find_all('p'):
        authors.append(d2.find('span',class_ = 'sc-1w3fpd7-0 dnCnAO').text.strip())
        date.append(d2.find('span',class_ = 'sc-1thf9ly-2 dvggWt').text.strip())
```

```python
paper_df = pd.DataFrame()
paper_df['Title'] = title
paper_df['Authors'] = authors
paper_df['Date'] = date
paper_df['url'] = url
```

9. Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location 
iv) Ratings
v) Image URL


```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/welcome-back')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
rest = soup.find_all('div',class_='restnt-main-wrap clearfix')
rest
```

```python
rest_name = []
cuisine = []
location = []
ratings = []
url = []
img_url = []
```

```python
for data in rest:
    for d1 in data.find_all('a'):
        url.append(d1.get('href'))
    rest_name.append(data.find('a',class_='restnt-name ellipsis').text.strip())
    location.append(data.find('div',class_='restnt-loc ellipsis').text.strip())
    ratings.append(data.find('div',class_='restnt-rating rating-4').text.strip())
    cuisine.append(data.find('span',class_='double-line-ellipsis').text.split('|')[1])
    img_url.append(data.find('img',class_='no-img').get('data-src'))
    
```

```python
final_table = pd.DataFrame()

final_table['Restaurant Name'] = rest_name
final_table['Cuisine'] = cuisine
final_table['Location'] = location
final_table['Ratings'] = ratings
final_table['Url'] = url
final_table['Image'] = img_url
```
