# 5) Deleting Columns And Encoding The Required Columns 

<br><font face = "TImes New Roman" size = "4"> <b> We will now be deleting the columns that are not important to our analysis. 

```python
DelColumns = ['Restaurant ID', 'Restaurant Name','Address','Country Code',
                'Locality', 'Locality Verbose','Cuisines','Longitude','Latitude',
                'Currency','Votes',
                "Aggregate rating","Switch to order menu",'Rating color']
MainZomato = dff.drop(DelColumns,axis=1)
MainZomato.head().reset_index(inplace=True)
MainZomato.drop(['index'], axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
labe=LabelEncoder()
```

```python
dfe=labe.fit_transform(MainZomato['City'])
pd.Series(dfe)
MainZomato['City']=dfe
MainZomato['Has Table booking']=dfe
MainZomato['Has Online delivery']=dfe
MainZomato['Is delivering now']=dfe
MainZomato['Rating text']=dfe
```
